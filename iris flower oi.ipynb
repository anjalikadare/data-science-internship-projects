```python
import pandas as pd # for data analysis
import numpy as np # for data analysis
import matplotlib.pyplot as plt # for data viz
import seaborn as sns # for data viz & loading data set
import warnings # ignoring some errors generated by seaborn
warnings. filterwarnings("ignore")
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.dtypes
iris.describe()
```

```python
sns.pairplot(iris, hue='species', palette='Dark2')
```

```python
plt.figure(figsize=(7,4))
sns.heatmap(iris.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
setosa = iris[iris['species']=='setosa']
setosa_sl = setosa['sepal_length']
setosa_sw = setosa['sepal_width']
sns.kdeplot(data=iris, x=setosa_sl, y=setosa_sw, cmap='plasma', shade=True, shade_lowest=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = iris.drop('species', axis=1)
y = iris['species']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
predictions = svc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(y_test, predictions)
print(a)
```
